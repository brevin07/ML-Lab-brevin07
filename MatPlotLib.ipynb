```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 1: Basic line plot of a sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 2: Simple scatter plot with random data
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure()
plt.scatter(x, y, c='blue', marker='o')
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 3: Bar chart of categorical data
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]
plt.figure()
plt.bar(categories, values, color='green')
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 4: Pie chart with an exploded slice
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice
plt.figure()
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title("Pie Chart")
plt.axis('equal')  # Ensures that the pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 5: Histogram of normally distributed data
data = np.random.randn(1000)
plt.figure()
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 6: Plotting sin(x) and cos(x) on the same figure
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure()
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title("Multiple Lines")
plt.xlabel("x")
plt.ylabel("Function value")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 7: Creating a 2x2 grid of subplots
x = np.linspace(0, 10, 100)
plt.figure()

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Subplot 1")

plt.subplot(2, 2, 2)
plt.scatter(x, np.sin(x), color='red')
plt.title("Subplot 2")

plt.subplot(2, 2, 3)
plt.bar(['a', 'b', 'c', 'd'], [np.random.randint(1, 10) for _ in range(4)])
plt.title("Subplot 3")

plt.subplot(2, 2, 4)
plt.hist(np.random.randn(100), bins=10)
plt.title("Subplot 4")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 8: Plot with grid lines
x = np.linspace(0, 10, 100)
y = np.exp(-x/3) * np.sin(x)
plt.figure()
plt.plot(x, y, marker='o')
plt.title("Exponential Decay with Sine Wave")
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 9: Logarithmic-style plot with custom ticks
x = np.linspace(0, 10, 100)
y = np.log1p(x)
plt.figure()
plt.plot(x, y, color='orange')
plt.title("Logarithmic Plot")
plt.xlabel("x value")
plt.ylabel("log(1+x)")
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.linspace(0, np.log1p(10), 5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 10: Sine and Cosine with different line styles and a legend
x = np.linspace(0, 2 * np.pi, 100)
plt.figure()
plt.plot(x, np.sin(x), 'r-', label='sin(x)')
plt.plot(x, np.cos(x), 'b--', label='cos(x)')
plt.title("Sine and Cosine with Legends")
plt.xlabel("Angle (radians)")
plt.ylabel("Value")
plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 11: Sine wave plotted with markers and a dashed line
x = np.linspace(0, 10, 50)
plt.figure()
plt.plot(x, np.sin(x), marker='x', linestyle='--', color='magenta')
plt.title("Sine Wave with Markers")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 12: Plot with an annotation indicating a peak
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.title("Annotated Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
# Annotate the peak near pi/2
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 0.8),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 13: Plot with logarithmic scales on both axes
x = np.linspace(1, 100, 100)
y = np.exp(x / 20)
plt.figure()
plt.plot(x, y)
plt.xscale('log')
plt.yscale('log')
plt.title("Log-Log Plot")
plt.xlabel("x (log scale)")
plt.ylabel("y (log scale)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 14: Plot demonstrating error bars
x = np.arange(0, 10, 1)
y = np.sin(x)
yerr = np.random.uniform(0.1, 0.2, size=x.size)
plt.figure()
plt.errorbar(x, y, yerr=yerr, fmt='o', ecolor='black', capsize=5)
plt.title("Error Bar Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 15: Plot that fills the area between two curves
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5
plt.figure()
plt.plot(x, y1, 'b-', label='sin(x)')
plt.plot(x, y2, 'r-', label='sin(x) + 0.5')
plt.fill_between(x, y1, y2, color='gray', alpha=0.3)
plt.title("Fill Between Plot")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 17: Contour plot of a function over a grid
x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)
plt.figure()
cp = plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.title("Contour Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(cp)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 18: 3D surface plot using mpl_toolkits.mplot3d
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 19: Polar plot displaying a radial function
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(theta))
plt.figure()
plt.polar(theta, r)
plt.title("Polar Plot")
plt.show()
```
